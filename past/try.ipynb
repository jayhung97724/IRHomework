```python
# -*- coding: utf-8 -*-
import json
import jieba
import jieba.posseg as pseg
jieba.load_userdict('dict.txt.big.txt')
from collections import OrderedDict
import locale
import pandas as pd
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
with open('wiki_data_small.json', 'rb') as json_data:    
    wiki_data = json.load(json_data)
```

```python
top = 0
tempDict = OrderedDict()
for key in range(len(wiki_data)):
    #key = '0'
    words = pseg.cut(wiki_data[str(key)])
    for word in words:
        if not tempDict.has_key(word.word):
            tempDict.update({word.word:top})
            top += 1
    print key
print json.dumps(tempDict, encoding='utf-8', ensure_ascii=False)
```

```python
vec = []
for key in range(len(wiki_data)):
    temp_vec = [0 for j in range(top)]
    words = pseg.cut(wiki_data[str(key)])
    for word in words:
        temp_vec[tempDict[word.word]]+=1
    vec.append(temp_vec)
print vec
```

```python
from math import sqrt
def cosine(v1, v2):
    inner_product = sum([ v1[i]*v2[i] for i in range(len(v1))])
    length_v1 = sqrt(sum([ v1[i]*v1[i] for i in range(len(v1))]))
    length_v2 = sqrt(sum([ v2[i]*v2[i] for i in range(len(v2))]))
    cosine= inner_product/(length_v1*length_v2)
    return cosine
```

```python
def findAnswer(question):
    temp_vec = [0 for j in range(top)]
    temp_vecA = [0 for j in range(top)]
    temp_vecB = [0 for j in range(top)]
    temp_vecC = [0 for j in range(top)]
    words = pseg.cut(question['Question'])
    for word in words:
        temp_vec[tempDict[word.word]]+=1
    words = pseg.cut(question[' A'])
    for word in words:
        temp_vecA[tempDict[word.word]]+=1
    words = pseg.cut(question['B'])
    for word in words:
        temp_vecB[tempDict[word.word]]+=1
    words = pseg.cut(question['C'])
    for word in words:
        temp_vecC[tempDict[word.word]]+=1
    s1 = cosine(temp_vec, temp_vecA)
    s2 = cosine(temp_vec, temp_vecB)
    s3 = cosine(temp_vec, temp_vecC)
    if s1 > s2 and s1 > s3:
        return "A"
    elif s2 > s1 and s2 > s3:
        return "B"
    elif s3 > s1 and s3 > s2:
        return "c"
```
